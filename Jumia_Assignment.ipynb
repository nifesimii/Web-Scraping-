```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from fake_useragent import UserAgent
```

```python
url = 'https://www.jumia.com.ng/catalog/?q=balls'
ua = UserAgent()
userAgent = ua.random
headers = {'User-Agent': userAgent}
page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
cart_items = soup.find_all('a', class_="core")

names = []
prices = []
old_prices = []
ratings = []
img_sources = []
    
for item in cart_items: 
        #Image sources
    if item.find('img', class_="img") is not None:
        image = item.find('img', class_="img")
        img_sources.append(image.get('src'))
    else:
        img_sources.append(None)


    #prices
    if item.find('span', class_ = "curr") is not None:
        price = item.find('span', class_="curr")
        if price:
            prices.append(price.text.strip())
        else:
            prices.append(None) 
 
    #old_prices
    if item.find('span', class_ = "old") is not None:
        old_price = item.find('span', class_ = "old")
        old_prices.append(old_price.text)
    else:
        old_prices.append(None)
        
    #names
    if item.find('h2', attrs={'class': 'name'}) is not None:
        name = item.find('h2', attrs={'class': 'name'})
        names.append(name.text)
    else:
        names.append(None)


items_df = pd.DataFrame({
    'images': img_sources,
    'names': names,
    'prices': prices,
    'old_prices': old_prices
#     'rating': ratings
})

items_df
```

```python
#cleaning up the price information
def clean_price(x):
    if pd.isnull(x) or x is None:
        return None
    x = str(x).replace('', '').replace(',', '').strip()
    if '-' in x:
        x = x.split('-')[0]  # take the first part (minimum price)
    return int(float(x.strip()))  # <-- turn it into int

#adding other colums
items_df['prices'] = items_df['prices'].apply(clean_price)
items_df['old_prices'] = items_df['old_prices'].apply(clean_price)
items_df['discount'] = ((items_df['old_prices'] - items_df['prices']) / items_df['old_prices']) * 100

items_df



```

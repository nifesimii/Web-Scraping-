```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Venezuela'

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
import pandas as pd

# Assuming `soup` is already defined
matches = soup.find_all('a')  # find_all can't use CSS selectors like 'p > a' directly

data = []

for i in matches:
    if i.parent.name == 'p':  # Ensure <a> is directly inside a <p>
        data.append(i.get_text(strip=True))

df = pd.DataFrame(data, columns=['Text'])

print(df)
```

```python
import pandas as pd

# Assuming `soup` is already defined
match = soup.find('a')  # find returns a single element (or None)

data = []

if match or match.parent.name == 'p':  # Check that match exists and parent is <p>
    data.append(match.get_text(strip=True))

df = pd.DataFrame(data, columns=['Text'])

print(df)


```
